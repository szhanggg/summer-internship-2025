```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MODEL_NAMES = ["satfull", "unetfull", "sathalf", "unethalf", "satquarter", "unetquarter", "sateighth", "uneteighth"][::-1]
```

```python
data = {}
```

```python
datapath = "/explore/nobackup/projects/pix4dcloud/szhang16/test_results/"
for model in MODEL_NAMES:
    sample_data = np.load(datapath + f"{model}.npy")
    data[model] = sample_data
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_ylim([0, 0.55])
ax.boxplot(data.values())
ax.set_xticklabels(data.keys())
```
